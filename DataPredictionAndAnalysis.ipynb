## Task 2:Predictive Modeling: Linear Regression.

***(2.1): On the training set, train a linear regression model to predict the target feature, using only the  descriptive features selected in exercise (1) above***

**Since our target variable death_yn is categorical with values "Yes" or "No", it would be better to convert these into a binary format that can be used with logistic regression. Typically, this is done by encoding "Yes" as 1 and "No" as 0.**

```python
train_df['death_yn'] = train_df['death_yn'].map({'Yes': 1, 'No': 0})
X_train = train_df.drop('death_yn', axis=1)
y_train = train_df['death_yn']
X_train_dum = pd.get_dummies(X_train, drop_first=True)
linear_model = LinearRegression()
linear_model.fit(X_train_dum, y_train)
```

**We need to convert the categorical columns to give dummy values for the purpose of training them further**

```python
coefficients = pd.DataFrame(linear_model.coef_, index=X_train_dum.columns, columns=['Coefficient'])
print(coefficients)
```
